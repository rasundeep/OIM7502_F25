# More Matplotlib Functionality

```python
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import norm
import yfinance as yf
```

### Figures and Axes
* mpl object are these two objects
* this means you can create and access each 
* in general you must have a figure and then you can add axes (subplots)

```python

fig = plt.figure(figsize = (9,6))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(30).cumsum(),'g--')
ax1.hist(np.random.randn(100), bins=15, color='g', edgecolor='w', alpha = .5)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.random(30))

```

```python
fig, axes = plt.subplots(2,2,sharex=True, sharey=True, figsize=(8,6))
for i in range(2):
    for j in range(2):
      axes[i,j].hist(np.random.randn(500), bins=25, color=(.1,.4,.4), edgecolor='w')

plt.subplots_adjust(wspace=0.05,hspace=0.05)
```

# Line Graphs

```python
plt.plot(np.random.randn(30).cumsum(), "bo--")
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'b--', label='Default')
plt.plot(data, 'r', drawstyle='steps-pre', label= 'Steps pre', alpha=.7)
plt.legend(loc=9)
```

```python
data = yf.download('^GSPC', start= '2007-01-01', end= '2012-01-01', auto_adjust=False)['Close']
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

crisis = [
    (datetime(2007, 10, 11), 'Peak'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

data.plot(ax=ax, c='royalblue')
ax.set_ylim([600, 1800])
ax.grid(alpha=.2)

for date, label in crisis:
    ax.annotate(label,
                xy=(date, data.asof(date) +75),
                xytext=(date, data.asof(date) +75),
                arrowprops=dict(color='grey', headwidth=5, width=2, headlength=4),
                horizontalalignment='left',verticalalignment='top')

```
